```python
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
# Reading in the Data
data = pd.read_csv()

data.head()
```

```python
# Time series requires dates to be in order so converting and sorting the data
data["FlightDate"] = pd.to_datetime(data["FlightDate"])

data = data.sort_values("FlightDate")
data.set_index("FlightDate", inplace=True)
```

```python
daily_delay = data["ArrDelay"].resample("D").mean()

daily_delay = daily_delay.fillna(0)
```

```python
plt.figure(figsize=(10,5))
plt.plot(daily_delay)
plt.title("Average Daily Flight Delay")
plt.xlabel("Date")
plt.ylabel("Delay (minutes)")
plt.show()
```

```python
train_size = int(len(daily_delay) * 0.8)

train = daily_delay[:train_size]
test = daily_delay[train_size:]
```

```python
model = ARIMA(train, order=(2,1,2))

model_fit = model.fit()
```

```python
forecast = model_fit.forecast(steps=len(test))
```

```python
rmse = mean_squared_error(test, forecast, squared=False)

print("RMSE:", rmse)
```
