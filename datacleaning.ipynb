```python
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Airline_Delay_Cause.csv")
data
```

```python
data.columns.values
```

```python
df_baysian = data.drop(columns=['carrier', 'airport_name', 'nas_ct'])
df_baysian
```

```python
df_baysian.to_csv('cleanedbaysian.csv')
```

```python
scaler = MinMaxScaler(feature_range=(1, 5))

df_data = data[['carrier_name', 'airport', 'arr_delay']]
df_data = df_data.dropna()

# drop outliers
Q1 = df_data['arr_delay'].quantile(0.25)
Q3 = df_data['arr_delay'].quantile(0.75)
IQR = Q3 - Q1
df_range = df_data[(df_data['arr_delay'] >= Q1 - 1.5*IQR) & (df_data['arr_delay'] <= Q3 + 1.5*IQR)]
df_range = df_range[(df_range['arr_delay'] < 4000)]


df_cf = df_range.pivot_table(index='airport', columns='carrier_name', aggfunc='mean', values='arr_delay')
df_filled = df_cf.apply(lambda row: row.fillna(row.mean()), axis=1)


df_filled = pd.DataFrame(scaler.fit_transform(df_filled.values.reshape(-1, 1)).reshape(df_filled.shape),
                     index=df_filled.index, columns=df_filled.columns)

df_cf[df_cf.notna()] = df_filled
df_cf
```

```python
df_filled.to_csv('filledcf.csv')
df_cf.to_csv('cleanedcf.csv')
```
