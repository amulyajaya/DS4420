```python
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("Airline_Delay_Cause.csv")
data
```

```python
data.columns.values
```

```python
df_baysian = data.drop(columns=['carrier', 'airport_name', 'nas_ct'])
df_baysian
```

```python
df_baysian.to_csv('cleanedbaysian.csv')
```

```python
scaler = MinMaxScaler(feature_range=(1, 5))

df_columns = data[['carrier_name', 'airport', 'arr_delay']]
df_cleaned = df_columns.dropna()
df_cleaned['arr_delay'] = scaler.fit_transform(df_cleaned[['arr_delay']])
df_cf = df_cleaned.pivot_table(index='airport', columns='carrier_name', aggfunc='mean', values='arr_delay')
df_cf
```

```python
df_cf.to_csv('cleanedcf.csv')
```
